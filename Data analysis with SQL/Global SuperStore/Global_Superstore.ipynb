```python
import pandas as pd
import sqlite3
```

```python
# SQLite veritabanı bağlantısını oluşturun
conn = sqlite3.connect('superstore.db')
```

```python
# Veri setini okuyun ve SQLite veritabanına aktarın
data = pd.read_csv('/content/sample_data/superstore.csv')
data.to_sql('superstore', conn, if_exists='replace', index=False)
```

```python
# SQL sorguları

#Tabloya göz atalım:
query = "SELECT * from superstore"
tables = pd.read_sql(query, conn)
tables
```

```python
#Toplam Satış Mikatrı:
query1 = "SELECT Sum(Sales) as TotalSales from superstore"
tables1 = pd.read_sql(query1, conn)
tables1
```

```python
#Toplam Kar:
query2 = "SELECT Sum(Profit) as TotalProfit from superstore"
tables2 = pd.read_sql(query2, conn)
tables2
```

```python
#En çok satılan ürün:
query3 = "SELECT `Product.Name`, MAX(Quantity) as MaxQuantity from superstore Group by `Product.Name` order by MaxQuantity DESC LIMIT 1;"
tables3 = pd.read_sql(query3, conn)
tables3
```

```python
# Los Angeles Yapılan Satışlar:
query4 = "SELECT * from superstore where City = 'Los Angeles'"
tables4 = pd.read_sql(query4, conn)
tables4
```

```python
# Belirli Bir Kategoriye Ait Ürünler:
query5 = "SELECT * from superstore where Category = 'Technology'"
tables5 = pd.read_sql(query5, conn)
tables5
```

```python
# Bir Yıl İçindeki En Yüksek Karlı Ürün:
query6 = "SELECT `Product.Name`, Max(Profit) as MaxProfit from superstore where year = '2014' Group by `Product.Name` order by MaxProfit desc LIMIT 1;"
tables6 = pd.read_sql(query6, conn)
tables6
```

```python
# Bir Yıl İçindeki En Yüksek 10 Karlı Ürün:
query7 = "SELECT `Product.Name`, Max(Profit) as MaxProfit from superstore where year = '2014' Group by `Product.Name` order by MaxProfit desc LIMIT 10;"
tables7 = pd.read_sql(query7, conn)
tables7
```

```python
# Toplam Harcamaları En Yüksek Olan Müşteri:
query8 = "SELECT `Customer.ID`,`Customer.Name`, sum(sales) as TotalSpent from superstore group by `Customer.ID`,`Customer.Name` order by TotalSpent DESC LIMIT 1;"
tables8 = pd.read_sql(query8, conn)
tables8
```

```python
# En Çok Satın Alma Yapan 10 Müşteri
query9 = "Select `Customer.ID`, `Customer.Name`, Count(`Order.ID`) as TotalOrders from superstore group by `Customer.ID`, `Customer.Name` order by TotalOrders DESC LIMIT 10;"
tables9 = pd.read_sql(query9, conn)
tables9
```

```python
# Toplam harcaması en yüksek olan Sean Miller'in tüm alımlarına bakalım
query10 = "Select * from superstore where `Customer.Name` = 'Sean Miller'"
tables10 = pd.read_sql(query10, conn)
tables10
```

```python
# En fazla sipariş veren müşteri
query11 = "Select `Customer.ID`, `Customer.Name`, Count(DISTINCT `Order.ID`) as TotalOrders from superstore group by `Customer.ID`, `Customer.Name` order by TotalOrders DESC LIMIT 10;"
tables11 = pd.read_sql(query11, conn)
tables11
```

```python
# En Yüksek Tek Seferlik Harcama Yapan Müşteri:
query12 = "Select `Customer.ID`, `Customer.Name`, max(Sales) as MaxSinglePurchase from superstore group by `Customer.ID`, `Customer.Name` order by MaxSinglePurchase DESC LIMIT 1;"
tables12 = pd.read_sql(query12, conn)
tables12
```

```python
# Ortalama Harcamaları Sıralama
query13 = "Select `Customer.ID`, `Customer.Name`, AVG(sales) as AverageSpent from superstore group by `Customer.ID`, `Customer.Name` order by AverageSpent DESC"
tables13 = pd.read_sql(query13, conn)
tables13
```

```python
# Toplam Satışlara Göre Ürünlerin Sıralanması
query14 = "Select `Product.ID`, `Product.Name`, SUM(quantity) as TotalSales from superstore group by `Product.ID`, `Product.Name` order by TotalSales DESC"
tables14 = pd.read_sql(query14, conn)
tables14
```

```python
#En Yüksek Kar Marjına Sahip Ürünlerin Listesi:
query15 = "Select `Product.ID`, `Product.Name`, avg(profit) as AverageProfit from superstore group by `Product.ID`, `Product.Name` order by AverageProfit DESC"
tables15 = pd.read_sql(query15, conn)
tables15
```

```python
# Belirli Bir Kategorideki En Pahalı Ürün
query16 = "Select `Product.ID`, `Product.Name`, MAX(sales) as HighestPrice from superstore where category = 'Technology' group by `Product.ID`, `Product.Name`"
tables16 = pd.read_sql(query16, conn)
tables16
```

```python
# En maliyetli 10 ürün
query17 = "Select `Product.ID`, `Product.Name`, MAX(`Shipping.Cost`) as HighestCost from superstore group by `Product.ID`, `Product.Name` order by HighestCost DESC LIMIT 10;"
tables17 = pd.read_sql(query17, conn)
tables17
```

```python
# Bölgelere Göre Toplam Satışlar
query18 = "Select `Region` , sum(quantity) as TotalSales from superstore group by `Region` order by TotalSales DESC;"
tables18 = pd.read_sql(query18, conn)
tables18
```

```python
# Ülkelere Göre En Çok Sipariş Veren Müşteriler
query19 = "Select country, `Customer.Name`, Count(`Order.ID`) as OrderCount from superstore group by country, `Customer.Name` order by OrderCount DESC;"
tables19 = pd.read_sql(query19, conn)
tables19
```

```python
# Şehirlere Göre Kar Marjı
query20 = "Select city, AVG(profit) as AverageProfit from superstore group by city order by AverageProfit DESC;"
tables20 = pd.read_sql(query20, conn)
tables20
```

```python
# Belirli Bir Bölgedeki Toplam Satışlar ve Karlılık
query21 = "Select `Region`, SUM(quantity) as TotalSales, AVG(Profit) as AverageProfit from superstore where `Region` = 'East' group by `Region`"
tables21 = pd.read_sql(query21, conn)
tables21
```
