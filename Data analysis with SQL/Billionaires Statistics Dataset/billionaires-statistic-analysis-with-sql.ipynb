```python
import pandas as pd
import sqlite3
```

```python
# SQLite veritabanı bağlantısını oluşturun
conn = sqlite3.connect('billionaires.db')  
```

```python
# Veri setini okuyun ve SQLite veritabanına aktarın
data = pd.read_csv('/kaggle/input/billionaires-statistics-dataset/Billionaires Statistics Dataset.csv')
data.to_sql('billionaires', conn, if_exists='replace', index=False)
```

```python
# SQL sorguları

#Tüm tabloya göz atalım
query = "Select * From billionaires"
tables = pd.read_sql(query, conn)
tables
```

```python
#Top 10 en yüksek net varlığa sahip milyarder
query1 = "Select personName, finalWorth from billionaires order by finalWorth DESC;"
tables1 = pd.read_sql(query1, conn)
tables1
```

```python
# Ortalama milyarder yaşı
query2 = "Select avg(age) as AverageAge FROM billionaires"
tables2 = pd.read_sql(query2, conn)
tables2
```

```python
# Kategorilere göre milyarder sayıları
query3 = "Select category, Count(*) as TotalBillionaires from billionaires group by category order by TotalBillionaires DESC;"
tables3 = pd.read_sql(query3, conn)
tables3
```

```python
# Top 10 milyarderin servetlerinin kaynağı
query4 = "Select personName, source from billionaires order by finalWorth DESC LIMIT 10;"
tables4 = pd.read_sql(query4, conn)
tables4
```

```python
# En fazla milyarder sayısına sahip 5 ülke
query5 = "Select country, count(*) as TotalBillionaires from billionaires group by country order by TotalBillionaires DESC LIMIT 5;"
tables5 = pd.read_sql(query5, conn)
tables5
```

```python
# Cinsiyete göre milyarderlerin ortalama nihai değeri
query6 = "Select gender, AVG(finalWorth) as AverageWorth from billionaires group by gender;"
tables6 = pd.read_sql(query6, conn)
tables6
```

```python
#En genç milyarder kim?
query7 = "Select personName, age, SOURCE from billionaires where age IS NOT NULL order by age ASC LIMIT 1;"
tables7 = pd.read_sql(query7, conn)
tables7
```

```python
# Her ülkedeki milyarderlerin toplam net değerini hesapla
query8 = "Select country, sum(finalWorth) as TotalWorth from bilLionaires group by country order by TotalWorth DESC;"
tables8 = pd.read_sql(query8, conn)
tables8
```

```python
# Kaynağa göre milyarder sayısı
query9 = "Select source, Count(*) as CountOfBillionaires from billionaires group by source order by CountOfBillionaires DESC;"
tables9 = pd.read_sql(query9, conn)
tables9
```

```python
# Milyarderlerin cinsiyete göre dağılımı
query10 = "Select gender, count(*) as TotalBillionaires from billionaires group by gender;"
tables10 = pd.read_sql(query10, conn)
tables10
```

```python
# Milyarderlerin servet kaynaklarının her kategorideki dağılımı
query11 = "Select category, source, count(*) as TotalBillionaires from billionaires group by category, source order by TotalBillionaires DESC;"
tables11 = pd.read_sql(query11, conn)
tables11
```

```python
# Milyarderlerin farklı ülke ve kategorilerdeki dağılımı
query12 = "Select category, country, count(*) as TotalBillionaires from billionaires group by category, country order by TotalBillionaires DESC;"
tables12 = pd.read_sql(query12, conn)
tables12
```

```python
# Kendi kendine milyarder olanlarla miras alınan zenginlikler
query13 = "Select status, count(*) as TotalBillionaires from billionaires group by status;"
tables13 = pd.read_sql(query13, conn)
tables13
```
