```python
df['Streams (Billions)']=df['Streams (Billions)'].str.replace(',','')
df['Streams (Billions)']=df['Streams (Billions)'].astype('int')
```

```python
df['Release Date']=df['Release Date'].astype('datetime64[ns]')
```

```python
df.info()
```

```python
art=df[['Artist','Streams (Billions)']].groupby('Artist').sum()
art=art.sort_values(by='Streams (Billions)',ascending=False)
art=art.head(10)
```

```python
plt.style.use('seaborn-dark-palette')
art.plot(kind='bar',figsize=(10,6),title='Top 5 artists by Streams count ',edgecolor ='black')
```

```python
# Select the "Song" and "Streams (Billions)" columns
songs = df["Song"].head(10)
streams = df["Streams (Billions)"].head(10)

# Create a figure with a larger size
plt.figure(figsize=(20, 6))

# Create a bar chart
plt.barh(songs, streams)

# Add axis labels and a title
plt.xlabel("Streams (Billions)")
plt.ylabel("Song")
plt.title("Top 10 Spotify Songs")

# Show the plot
plt.show()
```

```python
# Convert the "Release Date" column to a datetime data type
df["Release Date"] = pd.to_datetime(df["Release Date"])
```

```python
df["Year"] = df["Release Date"].dt.year
```

```python
# Group the data by year and compute the total number of streams for each year
yearly_streams = df.groupby("Year")["Streams (Billions)"].sum()
```

```python
# Set the figure size
plt.figure(figsize=(10, 6))
# Create a line chart of the yearly streams
yearly_streams.plot()


# Add axis labels and a title
plt.xlabel("Year")
plt.ylabel("Total Streams (Billions)")
plt.title("Yearly Streams of Top 100 Spotify Songs")

# Show the plot
plt.show()
```

```python
# Select the "Song" and "Streams (Billions)" columns
songs = df["Song"].head(10)
streams = df["Streams (Billions)"].head(10)

# Set the figure size
plt.figure(figsize=(20, 6))

# Create a bar chart
plt.bar(songs, streams)

# Add horizontal and vertical grid lines
plt.grid(True, axis="both")

# Add axis labels and a title
plt.xlabel("Song")
plt.ylabel("Streams (Billions)")
plt.title("Top 100 Spotify Songs")

# Show the plot
plt.show()
```

```python
# Select the "Artist" and "Streams (Billions)" columns
artists = df["Artist"].head(10)
streams = df["Streams (Billions)"].head(10)

# Set the figure size
plt.figure(figsize=(20, 6))

# Create a bar chart
plt.bar(artists, streams)

# Add data labels to the bars
for x, y in zip(artists, streams):
    plt.text(x, y, str(y), ha="center", va="bottom")

# Add a legend
plt.legend(["Total Streams"])

# Add axis labels and a title
plt.xlabel("Artist")
plt.ylabel("Streams (Billions)")
plt.title("Top 100 Spotify Songs")

# Show the plot
plt.show()
```

```python

# Select the "Song" and "Streams (Billions)" columns
songs = df["Song"].head(10)
streams = df["Streams (Billions)"].head(10)

# Create a pie chart
plt.pie(streams, labels=songs, autopct="%1.1f%%")

# Add a title
plt.title("Top 10 Spotify Songs")

# Show the plot
plt.show()
```

This will create a pie chart with the top 10 songs, where the size of each slice is proportional to the number of streams for each song, and the label for each slice shows the percentage of the total streams it represents.
