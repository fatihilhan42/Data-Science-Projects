```python
import folium
import pandas as pd

# Load your CSV data
data = pd.read_csv('/content/istasyon_konum.csv', delimiter=';')

# Clean and convert 'enlem' and 'boylam' columns to float
data['enlem'] = data['enlem'].str.replace('.', '').astype(float) / 1000000.0
data['boylam'] = data['boylam'].str.replace('.', '').astype(float) / 1000000.0

# Create a map centered around Konya, Turkey
konya_map = folium.Map(location=[37.8716, 32.4846], zoom_start=13)

# Create a FeatureGroup to hold the markers
marker_group = folium.FeatureGroup(name='Bicycle Stations')

# Loop through the rows of your data and add markers for each station
for index, row in data.iterrows():
    station_name = row['station']
    latitude = row['enlem']
    longitude = row['boylam']
    region = row['bolge']
    platforms = row['peron_adet']

    # Create a marker for each station and add it to the FeatureGroup
    folium.Marker(
        location=[latitude, longitude],
        popup=f"Station: {station_name}<br>Region: {region}<br>Platforms: {platforms}",
    ).add_to(marker_group)

# Add the FeatureGroup with all markers to the map
marker_group.add_to(konya_map)

# Add layer control to toggle the markers on/off
folium.LayerControl().add_to(konya_map)

# Save the map as an HTML file
konya_map.save('bicycle_stations_map.html')

# Optionally, you can display the map in a Jupyter Notebook
konya_map
```
