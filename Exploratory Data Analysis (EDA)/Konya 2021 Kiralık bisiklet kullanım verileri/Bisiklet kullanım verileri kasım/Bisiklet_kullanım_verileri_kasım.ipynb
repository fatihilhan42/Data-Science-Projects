```python
import pandas as pd

# Kasım 2021 verilerinizin CSV dosyasının adını/yerini belirtin
kasim_veri_yolu = '/content/sample_data/kasım_data.csv'

# CSV dosyasını doğru sınırlayıcıyla (bu durumda noktalı virgül) okuyun
kasim_verisi = pd.read_csv(kasim_veri_yolu, delimiter=';')

# Temel bilgileri görüntüleyin
print("Kasım 2021 verileri hakkında temel bilgiler:")
print("Satır sayısı:", len(kasim_verisi))
print("Sütunlar:", kasim_verisi.columns)
print("Sütunların veri türleri:")
print(kasim_verisi.dtypes)
print("Özet istatistikler:")
print(kasim_verisi.describe())





```

```python
# Kasım verisi için 'rental_start_date' ve 'rental_end_date' sütunlarını datetime veri türüne çevirme
kasim_verisi['rental_start_date'] = pd.to_datetime(kasim_verisi['rental_start_date'], format='%d.%m.%Y %H:%M')
kasim_verisi['rental_end_date'] = pd.to_datetime(kasim_verisi['rental_end_date'], format='%d.%m.%Y %H:%M')

# Güncellenen sütunların veri tiplerini görüntüleme
print("Güncellenen sütunların veri tipleri:")
print(kasim_verisi.dtypes)

# Değişiklikleri doğrulamak için ilk birkaç satırı görüntüleme
print("\nDönüştürmeden sonraki ilk birkaç satır:")
print(kasim_verisi.head())
```

```python
# Calculate the 'duration' column in minutes for November data
kasim_verisi['duration'] = (kasim_verisi['rental_end_date'] - kasim_verisi['rental_start_date']).dt.total_seconds() / 60

# Display the first few rows to verify the changes
print("\n'duration' sütununu güncelledikten sonraki ilk birkaç satır:")
print(kasim_verisi.head())

# Save the updated data to a new CSV file
güncellenmiş_dosya_yolu = '/content/sample_data/güncellenmiş_kasim_verisi.csv'
kasim_verisi.to_csv(güncellenmiş_dosya_yolu, index=False, sep=';')
print("\nGüncellenmiş veri '{}' dosyasına kaydedildi".format(güncellenmiş_dosya_yolu))
```

```python
# Kasım verisi için enlem ve boylam değerlerini ayırma ve float veri türüne çevirme
kasim_verisi['start_latitude'], kasim_verisi['start_longitude'] = kasim_verisi['starting_location'].str.split(',').str
kasim_verisi['end_latitude'], kasim_verisi['end_longitude'] = kasim_verisi['ending_location'].str.split(',').str
kasim_verisi['start_latitude'] = kasim_verisi['start_latitude'].astype(float)
kasim_verisi['start_longitude'] = kasim_verisi['start_longitude'].astype(float)
kasim_verisi['end_latitude'] = kasim_verisi['end_latitude'].astype(float)
kasim_verisi['end_longitude'] = kasim_verisi['end_longitude'].astype(float)

# Değişiklikleri doğrulamak için ilk birkaç satırı görüntüleme
print("\nEnlem ve boylam sütunlarını güncelledikten sonraki ilk birkaç satır:")
print(kasim_verisi.head())

# Güncellenmiş veriyi yeni bir CSV dosyasına kaydetme
güncellenmiş_dosya_yolu = '/content/sample_data/güncellenmiş_kasim_verisi.csv'
kasim_verisi.to_csv(güncellenmiş_dosya_yolu, index=False, sep=';')
print("\nGüncellenmiş veri '{}' dosyasına kaydedildi".format(güncellenmiş_dosya_yolu))
```

```python
import pandas as pd
import folium
from folium.plugins import HeatMap

# 'updated_november_data.csv' dosyasını kullanarak veriyi okuyoruz. Bu CSV dosyası, enlem ve boylam sütunlarını içeriyor.
file_path = '/content/sample_data/güncellenmiş_kasim_verisi.csv'
november_data = pd.read_csv(file_path, delimiter=';')

# 'rental_start_date' ve 'rental_end_date' sütunlarını tarih saat veri türüne çeviriyoruz.
november_data['rental_start_date'] = pd.to_datetime(november_data['rental_start_date'], format='%Y-%m-%d %H:%M:%S')
november_data['rental_end_date'] = pd.to_datetime(november_data['rental_end_date'], format='%Y-%m-%d %H:%M:%S')

# 'starting_location' sütununu 'start_latitude' ve 'start_longitude' olarak ayırıyoruz.
november_data[['start_latitude', 'start_longitude']] = november_data['starting_location'].str.split(',', expand=True)

# 'ending_location' sütununu 'end_latitude' ve 'end_longitude' olarak ayırıyoruz.
november_data[['end_latitude', 'end_longitude']] = november_data['ending_location'].str.split(',', expand=True)

# Enlem ve boylam sütunlarını float veri türüne çeviriyoruz.
november_data['start_latitude'] = november_data['start_latitude'].astype(float)
november_data['start_longitude'] = november_data['start_longitude'].astype(float)
november_data['end_latitude'] = november_data['end_latitude'].astype(float)
november_data['end_longitude'] = november_data['end_longitude'].astype(float)

# 'rental_start_date' sütunundan sadece tarih kısmını içeren yeni bir sütun ('rental_date') oluşturuyoruz.
november_data['rental_date'] = november_data['rental_start_date'].dt.date

# Veriyi tarihe göre gruplayarak her günün kiralama sayısını hesaplıyoruz.
daily_rides = november_data.groupby('rental_date').size().reset_index(name='ride_count')

# Konya, Türkiye'nin koordinatlarına (enlem ve boylam) odaklanan bir Folium haritası oluşturuyoruz. İhtiyaca göre koordinatları ayarlayabilirsiniz.
konya_map = folium.Map(location=[38.0225067, 32.5125617], zoom_start=13)

# Günlük kiralama verilerini bir liste içinde saklıyoruz.
ride_data = november_data[['start_latitude', 'start_longitude']].values.tolist()

# Kiralama verileri ile bir Isı Haritası (HeatMap) katmanı oluşturuyoruz.
HeatMap(ride_data, radius=15).add_to(konya_map)

# Haritayı görüntülüyoruz.
konya_map
```
